```python
y=df.Type.values
x=df.drop('Type',axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
from keras.callbacks import EarlyStopping
```

```python
X_train.shape,y_train.shape
```

```python
model=Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
#OVERFIT EDIYORUM
model.fit(X_train,y_train,epochs=700, validation_data=(X_test,y_test),
         verbose=1)
```

```python
modelLoss=pd.DataFrame(model.history.history)
```

```python
modelLoss.head()
```

```python
modelLoss.plot(); #OVERFIT
```

```python
model=Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')

earlyStopping=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)

model.fit(X_train,y_train,epochs=700,
         validation_data=(X_test,y_test),verbose=2,
         callbacks=[earlyStopping])
```

```python
modelLoss=pd.DataFrame(model.history.history)
modelLoss.plot();
```

```python
model=Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.6))

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.6))

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.6))


model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')

earlyStopping=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)

model.fit(X_train,y_train,epochs=700,
         validation_data=(X_test,y_test),verbose=2,
         callbacks=[earlyStopping])
```

```python
lossDF=pd.DataFrame(model.history.history)
```

```python
lossDF.plot()
```

```python
predicts=model.predict_classes(X_test)
```

```python
predicts
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predicts))
```

```python
print(confusion_matrix(y_test,predicts))
```
